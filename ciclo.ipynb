```python
from pylab import *
from io import StringIO
%matplotlib inline
```

```python
s = open("pressione_1.txt").read()
t,P = transpose(loadtxt(StringIO(s)))

s = open("temperatura_1.txt").read()
t,T = transpose(loadtxt(StringIO(s)))

s = open("posizione_1.txt").read()
t,L = transpose(loadtxt(StringIO(s)))

T = T+273.15
P = P*1000
V = L*(0.0325/2)**2*3.14

print(P)
print(T)
print(L)
```

```python
m = 115.3
m = m/1000
```

```python
dh = (max(L[P<(0.5*1000)])-min(L[P<(0.5*1000)]))
print(dh)
DV = dh * (32.5/2000)**2 * pi
deltaV = (32.5/2000)**2*pi*(0.00001)
print("deltaV = ", deltaV)
DP = (max(P[P<(1.75*1000)])-min(P[P<(1.75*1000)]))
area0 = DV*DP
print("Area del ciclo: ", area0, "+/-", sqrt(deltaV**2+0.00001**2))

Lavoro = m * 9.81 * dh
dL = Lavoro*sqrt((0.0001/m)**2+(0.00001/dh)**2)
print("Lavoro: ", Lavoro, "+/- ", dL)
```
